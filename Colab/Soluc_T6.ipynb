```python
from IPython.display import clear_output

# Listas para armazenar os dados dos alunos
nomes_alunos = []  # Lista para armazenar os nomes dos alunos
numeros_alunos = []  # Lista para armazenar os números dos alunos
moradas_alunos = []  # Lista para armazenar as moradas dos alunos
bi_cc_alunos = []  # Lista para armazenar os números de Bilhete de Identidade/Cartão de Cidadão dos alunos
telefones_alunos = []  # Lista para armazenar os números de telefone dos alunos
emails_alunos = []  # Lista para armazenar os e-mails dos alunos
indice_aluno = 1  # Índice do próximo aluno a ser adicionado

# Funções lambda para capitalizar a primeira letra do nome, colocar a cidade em maiúsculas
# e converter os e-mails para minúsculas
capitalizar_primeira_letra = lambda nome: nome.capitalize()
cidade_maiusculas = lambda cidade: cidade.upper()
emails_minusculas = lambda email: email.lower()

def atualizar_numeros_alunos():
    global numeros_alunos, indice_aluno
    numeros_alunos = list(range(1, indice_aluno))
    indice_aluno = len(numeros_alunos) + 1

def listar_turma():
    if not nomes_alunos:
        print("A turma está vazia.")
    else:
        print("Lista de Alunos:")
        for indice, (nome, numero, morada, bi_cc, telefone, email) in enumerate(zip(nomes_alunos, numeros_alunos, moradas_alunos, bi_cc_alunos, telefones_alunos, emails_alunos), start=1):
            # Aplicar as funções lambda aos dados antes de exibir
            nome = capitalizar_primeira_letra(nome)
            morada = cidade_maiusculas(morada)
            email = emails_minusculas(email)
            # Imprime a lista de alunos enumerando-os
            print(f"Aluno {indice}: {nome} (Número: {numero}), Morada: {morada}, Bi/CC: {bi_cc}, Telefone: {telefone}, E-Mail: {email}")

def verificar_bi_cc_duplicado(bi_cc):
    if bi_cc in bi_cc_alunos:
        return True
    return False

while True:
    print("Menu:")
    print("1. Adicionar aluno")
    print("2. Remover aluno")
    print("3. Consultar turma")
    print("4. Atualizar dados do aluno")
    print("5. Limpar tela")
    print("6. Sair")

    escolha = input("Escolha uma opção (1-6): ")

    if escolha == "1":
        # Adicionar aluno
        while True:
            nome = input("Insira o nome do aluno: ")
            morada = input("Insira a morada do aluno: ")
            while True:
                bi_cc = input("Insira o número do Bilhete de Identidade/Cartão de Cidadão do aluno: ")
                if verificar_bi_cc_duplicado(bi_cc):
                    print("Erro: Número de Bi/CC já existente. Tente novamente.")
                else:
                    break
            telefone = input("Insira o número de telefone do aluno: ")
            email = input("Insira o e-mail do aluno: ")

            # Aplicar as funções lambda aos dados antes de adicionar à lista
            nome = capitalizar_primeira_letra(nome)
            morada = cidade_maiusculas(morada)
            email = emails_minusculas(email)

            # O número do aluno é automaticamente incrementado
            numero = indice_aluno
            indice_aluno += 1

            nomes_alunos.append(nome)
            moradas_alunos.append(morada)
            bi_cc_alunos.append(bi_cc)
            telefones_alunos.append(telefone)
            emails_alunos.append(email)
            numeros_alunos.append(numero)
            print("Aluno adicionado com sucesso!")

            continuar = input("Deseja continuar a adicionar alunos? (s/n): ")
            if continuar.lower() != 's':
                break

    elif escolha == "2":
        # Remover aluno
        listar_turma()
        if not nomes_alunos:
            print("Não existem alunos na turma.")
        else:
            numero = int(input("Insira o número do aluno que deseja remover: "))
            if numero in numeros_alunos:
                # Encontra o índice do número na lista dos números dos alunos
                indice = numeros_alunos.index(numero)
                # Remove todos os dados do aluno correspondente nas listas
                del nomes_alunos[indice]
                del moradas_alunos[indice]
                del bi_cc_alunos[indice]
                del telefones_alunos[indice]
                del emails_alunos[indice]
                del numeros_alunos[indice]
                print("Aluno removido com sucesso!")

                # Atualiza os números dos alunos após a remoção
                atualizar_numeros_alunos()

            else:
                print("Aluno não encontrado.")

    elif escolha == "3":
        # Consultar turma
        listar_turma()

    elif escolha == "4":
        # Atualizar dados do aluno
        listar_turma()
        if not nomes_alunos:
            print("A turma está vazia. Não há alunos para atualizar os dados.")
        else:
            numero_aluno = int(input("Insira o número do aluno cujos dados deseja atualizar: "))
            if numero_aluno in numeros_alunos:
                indice = numeros_alunos.index(numero_aluno)
                print("Escolha o dado que deseja atualizar:")
                print("1. Nome")
                print("2. Morada")
                print("3. Número de Bilhete de Identidade/Cartão de Cidadão")
                print("4. Número de Telefone")
                print("5. E-mail")
                opcao = input("Escolha uma opção (1-5): ")
                if opcao == "1":
                    novo_nome = input("Insira o novo nome do aluno: ")
                    nomes_alunos[indice] = capitalizar_primeira_letra(novo_nome)
                    print("Nome do aluno atualizado com sucesso.")
                elif opcao == "2":
                    nova_morada = input("Insira a nova morada do aluno: ")
                    moradas_alunos[indice] = cidade_maiusculas(nova_morada)
                    print("Morada do aluno atualizada com sucesso.")
                elif opcao == "3":
                    novo_bi_cc = input("Insira o novo número de Bilhete de Identidade/Cartão de Cidadão do aluno: ")
                    if verificar_bi_cc_duplicado(novo_bi_cc):
                        print("Erro: Número de Bi/CC já existente. Tente novamente.")
                    else:
                        bi_cc_alunos[indice] = novo_bi_cc
                        print("Número de Bilhete de Identidade/Cartão de Cidadão do aluno atualizado com sucesso.")
                elif opcao == "4":
                    novo_telefone = input("Insira o novo número de telefone do aluno: ")
                    telefones_alunos[indice] = novo_telefone
                    print("Número de telefone do aluno atualizado com sucesso.")
                elif opcao == "5":
                    novo_email = input("Insira o novo e-mail do aluno: ")
                    emails_alunos[indice] = emails_minusculas(novo_email)
                    print("E-mail do aluno atualizado com sucesso.")
                else:
                    print("Opção inválida. Escolha uma opção de 1 a 5.")
            else:
                print("Aluno não encontrado.")

    elif escolha == "5":
        # Código para limpar a tela (Google Colab)
        clear_output()

    elif escolha == "6":
        print("Programa Terminado. Obrigado!")
        break

    else:
        print("Opção inválida. Escolha uma opção de 1 a 6.")
```
