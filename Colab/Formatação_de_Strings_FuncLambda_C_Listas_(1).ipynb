```python
# Lista para armazenar as informações das pessoas
pessoas = []
id_counter = 1  # Contador para atribuir IDs únicos às pessoas

# Função lambda para indicar a idade com sufixo
idade_suf = lambda idade: "{} anos".format(idade) if idade != 1 else "{} ano".format(idade)

# Função lambda para colocar todas as letras do nome maiúsculas
nome_maiusc = lambda nome: nome.upper()

# Função lambda para colocar apenas a primeira letra da cidade Maiúscula
cidade_maius = lambda cidade: cidade.capitalize()

# Função para registar uma nova pessoa
def registar_pessoa():
    global id_counter
    #Aqui colocamos o contador global para as variáveis da lista
    print("Registando nova pessoa:")
    nome = input("Insira o nome da pessoa: ")
    idade = int(input("Insira a idade da pessoa: "))
    cidade = input("Insira a cidade da pessoa: ")

    #como temos o id que é automaticamente atualizado após o registo de uma nova
    #pessoa, este id passa a ser um indexador na lista, e como pode haver duas
    #ou mais pessoas com o mesmo nome, idade, etc. o id é sempre o fator de difernciação
    pessoa = {'id': id_counter, 'nome': nome_maiusc(nome), 'idade': idade_suf(idade), 'cidade': cidade_maius(cidade)}
    pessoas.append(pessoa)
    #nesta linha após a adição de cada pessoa e dos respetivos dados
    #o contador de id's é atualizado
    id_counter += 1
    print("Pessoa registada com sucesso!")

# Função para atualizar os dados de uma pessoa
def atualizar_pessoa():
    pessoa_id = int(input("Insira o ID da pessoa a ser atualizada: "))

    for i, pessoa in enumerate(pessoas):
        if pessoa['id'] == pessoa_id:
            print("A atualizar os dados para {}:".format(pessoa['nome']))
            opcao = input("Escolha o que deseja atualizar (1. Nome / 2. Idade / 3. Cidade): ")

            if opcao == "1":
                novo_nome = input("Insira o novo nome da pessoa: ")
                pessoas[i]['nome'] = nome_maiusc(novo_nome)
            elif opcao == "2":
                nova_idade = float(input("Insira a nova idade da pessoa: "))
                pessoas[i]['idade'] = idade_suf(nova_idade)
            elif opcao == "3":
                nova_cidade = input("Insira a nova cidade da pessoa: ")
                pessoas[i]['cidade'] = cidade_maius(nova_cidade)
            else:
                print("Opção inválida.")
                return

            print("Informações atualizadas com sucesso.")
            return

    print("ID {} não encontrado na lista.".format(pessoa_id))

# Função para excluir uma pessoa da lista
def excluir_pessoa():
    pessoa_id = int(input("Insira o ID da pessoa a ser excluída: "))

    for i, pessoa in enumerate(pessoas):
        if pessoa['id'] == pessoa_id:
            del pessoas[i]
            print("Pessoa removida com sucesso.")
            return

    print("ID {} não encontrado na lista.".format(pessoa_id))

# Função para consultar e exibir todas as pessoas registadas
def consultar_pessoas():
    if not pessoas:
        print("Não existem pessoas registadas na lista!")
    else:
        print("Pessoas registadas:")
        for pessoa in pessoas:
            print("ID: {}, Nome: {}, Idade: {}, Cidade: {}".format(pessoa['id'], pessoa['nome'], pessoa['idade'], pessoa['cidade']))

# Menu de opções
while True:
    print("\nMenu:") #o "\n" no texto dá um espaçamento
    print("1. Registar Pessoa")
    print("2. Atualizar Dados da Pessoa")
    print("3. Excluir Pessoa")
    print("4. Consultar Pessoas")
    print("5. Sair")

    opcao = input("Escolha uma opção (1/2/3/4/5): ")

    if opcao == "1":
        registar_pessoa()
    elif opcao == "2":
        atualizar_pessoa()
    elif opcao == "3":
        excluir_pessoa()
    elif opcao == "4":
        consultar_pessoas()
    elif opcao == "5":
        print("Programa terminado. Adeus e Obrigado!")
        break
    else:
        print("Opção inválida. Tente novamente.")




```
