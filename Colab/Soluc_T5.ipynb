```python
from IPython.display import clear_output

# Listas para armazenar os dados dos alunos
nomes_alunos = []  # Lista para armazenar os nomes dos alunos
numeros_alunos = []  # Lista para armazenar os números dos alunos
indice_aluno = 1  # Índice do próximo aluno a ser adicionado

def atualizar_numeros_alunos():
    global numeros_alunos, indice_aluno
    numeros_alunos = list(range(1, indice_aluno))
    indice_aluno = len(numeros_alunos) + 1

def listar_turma():
    if not nomes_alunos:
        print("A turma está vazia.")
    else:
        print("Lista de Alunos:")
        for indice, (nome, numero) in enumerate(zip(nomes_alunos, numeros_alunos), start=1):
            # Imprime a lista de alunos enumerando-os
            print(f"Aluno {indice}: {nome} (Número: {numero})")

while True:
    print("Menu:")
    print("1. Adicionar aluno")
    print("2. Remover aluno")
    print("3. Consultar turma")
    print("4. Alterar nome do aluno")
    print("5. Atualizar número do aluno")
    print("6. Limpar tela")
    print("7. Sair")

    escolha = input("Escolha uma opção (1-5): ")

    if escolha == "1":
        # Adicionar aluno
        nome = input("Insira o nome do aluno: ")

        # O número do aluno é automaticamente incrementado
        numero = indice_aluno
        indice_aluno += 1

        nomes_alunos.append(nome)
        numeros_alunos.append(numero)
        print("Aluno adicionado com sucesso!")

    elif escolha == "2":
        # Remover aluno
        if not nomes_alunos:
            print("Não existem alunos na turma.")
        else:
            # Antes de remover, listar a turma
            listar_turma()

            numero = int(input("Insira o número do aluno que deseja remover: "))
            if numero in numeros_alunos:
                # Encontra o índice do número na lista dos números dos alunos
                indice = numeros_alunos.index(numero)
                # Remove o nome e o número do aluno correspondente nas listas
                del nomes_alunos[indice]
                del numeros_alunos[indice]
                print("Aluno removido com sucesso!")

                # Atualiza os números dos alunos após a remoção
                atualizar_numeros_alunos()

            else:
                print("Aluno não encontrado.")

    elif escolha == "3":
        # Consultar turma
        listar_turma()

    elif escolha == "4":
        # Alterar nome do aluno
        if not nomes_alunos:
            print("A turma está vazia. Não há alunos para alterar.")
        else:
            numero = int(input("Insira o número do aluno cujo nome deseja alterar: "))
            if numero in numeros_alunos:
                # Encontra o índice do número na lista de números
                indice = numeros_alunos.index(numero)
                novo_nome = input("Insira o novo nome do aluno: ")
                # Atualiza o nome do aluno
                nomes_alunos[indice] = novo_nome
                print("Nome do aluno alterado com sucesso!")
            else:
                print("Aluno não encontrado.")

    elif escolha == "5":
        # Código para atualizar número do aluno
        if not nomes_alunos:
            print("A turma está vazia. Não há alunos para atualizar o número.")
        else:
            # Input para o utilizador introduzir o nº do Aluno a atualizar
            numero_antigo = int(input("Insira o número do aluno que deseja atualizar: "))
            if numero_antigo in numeros_alunos:
                # Encontra o índice do número na lista de números
                indice = numeros_alunos.index(numero_antigo)
                novo_numero = int(input("Insira o novo número do aluno: "))
                # Verifica se o novo número já existe
                if novo_numero in numeros_alunos:
                    print("Erro: Número de aluno já existente. Tente novamente.")
                else:
                    # Caso o nº não exista atualiza o nº de aluno
                    numeros_alunos[indice] = novo_numero
                    print("Número do aluno atualizado com sucesso!")
            else:
                print("Aluno não encontrado.")

    elif escolha == "6":
        # Código para limpar a tela (Google Colab)
        clear_output()

    elif escolha == "7":
        print("Programa Terminado! Obrigado!")
        break

    else:
        print("Opção inválida. Escolha uma opção de 1 a 7.")
```

```python
from IPython.display import clear_output

# Listas para armazenar os dados dos alunos
nomes_alunos = []       # Lista para armazenar os nomes dos alunos
numeros_alunos = []     # Lista para armazenar os números dos alunos
moradas_alunos = []     # Lista para armazenar as moradas dos alunos
bi_cc_alunos = []       # Lista para armazenar os números do BI/CC dos alunos
telefones_alunos = []   # Lista para armazenar os números de telefone dos alunos
emails_alunos = []      # Lista para armazenar os e-mails dos alunos

indice_aluno = 1  # Índice do próximo aluno a ser adicionado

# Funções lambda para formatação
formatar_maiuscula = lambda s: s.capitalize()
letras_minusculas = lambda s: s.lower()
adicionar_sufixo = lambda s, sufixo: s + sufixo

def atualizar_numeros_alunos():
    global numeros_alunos, indice_aluno
    numeros_alunos = list(range(1, indice_aluno))
    indice_aluno = len(numeros_alunos) + 1

def adicionar_aluno():
    global indice_aluno  # Declarando como global
    nome = formatar_maiuscula(input("Insira o nome do aluno: "))
    morada = letras_minusculas(input("Insira a morada do aluno: "))

    # Verificação para garantir que o BI/CC tenha 9 dígitos
    while True:
        bi_cc = adicionar_sufixo(input("Insira o número do BI/CC do aluno (9 dígitos): "), 'PT')
        if bi_cc.isdigit() and len(bi_cc) == 9:
            break
        else:
            print("Por favor, insira um número de BI/CC válido com 9 dígitos.")

    # Verificação para garantir que o telefone tenha 9 dígitos
    while True:
        telefone = input("Insira o número de telefone do aluno (9 dígitos): ")
        if telefone.isdigit() and len(telefone) == 9:
            break
        else:
            print("Por favor, insira um número de telefone válido com 9 dígitos.")

    # Verificação para garantir que o e-mail contenha pelo menos um '@'
    while True:
        email = letras_minusculas(input("Insira o e-mail do aluno: "))
        if '@' in email:
            break
        else:
            print("Por favor, insira um e-mail válido contendo pelo menos um '@'.")

    # O número do aluno é automaticamente incrementado
    numero = indice_aluno
    indice_aluno += 1

    nomes_alunos.append(nome)
    numeros_alunos.append(numero)
    moradas_alunos.append(morada)
    bi_cc_alunos.append(bi_cc)
    telefones_alunos.append(telefone)
    emails_alunos.append(email)

    print("Aluno adicionado com sucesso!")

while True:
    print("\nMenu:")
    print("1. Adicionar aluno")
    print("2. Remover aluno")
    print("3. Consultar turma")
    print("4. Alterar nome do aluno")
    print("5. Atualizar número do aluno")
    print("6. Limpar tela")
    print("7. Sair")

    escolha = input("Escolha uma opção (1-7): ")

    if escolha == "1":
        # Adicionar aluno
        adicionar_aluno()

        adicionar_outro = input("Deseja adicionar outro aluno? (s/n): ")
        if adicionar_outro.lower() != 's':
            break

    elif escolha == "2":
        # Remover aluno
        if not nomes_alunos:
            print("Não existem alunos na turma.")
        else:
            # Antes de remover, listar a turma
            listar_turma()

            numero = int(input("Insira o número do aluno que deseja remover: "))
            if numero in numeros_alunos:
                # Encontra o índice do número na lista dos números dos alunos
                indice = numeros_alunos.index(numero)
                # Remove o aluno correspondente nas listas
                del nomes_alunos[indice]
                del numeros_alunos[indice]
                del moradas_alunos[indice]
                del bi_cc_alunos[indice]
                del telefones_alunos[indice]
                del emails_alunos[indice]

                print("Aluno removido com sucesso!")

                # Atualiza os números dos alunos após a remoção
                atualizar_numeros_alunos()

            else:
                print("Aluno não encontrado.")

    elif escolha == "3":
        # Consultar turma
        listar_turma()

    elif escolha == "4":
        # Alterar nome do aluno
        if not nomes_alunos:
            print("A turma está vazia. Não há alunos para alterar.")
        else:
            numero = int(input("Insira o número do aluno cujo nome deseja alterar: "))
            if numero in numeros_alunos:
                # Encontra o índice do número na lista de números
                indice = numeros_alunos.index(numero)
                novo_nome = formatar_maiuscula(input("Insira o novo nome do aluno: "))
                # Atualiza o nome do aluno
                nomes_alunos[indice] = novo_nome
                print("Nome do aluno alterado com sucesso!")
            else:
                print("Aluno não encontrado.")

    elif escolha == "5":
        # Atualizar número do aluno
        if not nomes_alunos:
            print("A turma está vazia. Não há alunos para atualizar o número.")
        else:
            numero_antigo = int(input("Insira o número do aluno que deseja atualizar: "))
            if numero_antigo in numeros_alunos:
                indice = numeros_alunos.index(numero_antigo)
                novo_numero = int(input("Insira o novo número do aluno: "))
                if novo_numero in numeros_alunos:
                    print("Erro: Número de aluno já existente. Tente novamente.")
                else:
                    numeros_alunos[indice] = novo_numero
                    print("Número do aluno atualizado com sucesso!")
            else:
                print("Aluno não encontrado.")

    elif escolha == "6":
        # Limpar a tela (Google Colab)
        clear_output()

    elif escolha == "7":
        print("Programa Terminado! Obrigado!")
        break

    else:
        print("Opção inválida. Escolha uma opção de 1 a 7.")
```
