```python
#Vamos chamar a função com a sintaxe adaptada para o Colab (ambiente em Jupyter)
from IPython.display import clear_output


# Listas para armazenar os dados dos alunos
nomes_alunos = []  # Lista para armazenar os nomes dos alunos
numeros_alunos = []  # Lista para armazenar os números dos alunos
indice_aluno = 1  # Índice do próximo aluno a ser adicionado

while True:
    print("Menu:")
    print("1. Adicionar aluno")
    print("2. Remover aluno")
    print("3. Consultar turma")
    print("4. Alterar nome do aluno")
    print("5. Atualizar número do aluno")
    print("6. Limpar tela")
    print("7. Sair")

    escolha = input("Escolha uma opção (1-5): ")

    if escolha == "1":
        # Adicionar aluno
        nome = input("Insira o nome do aluno: ")

        # O número do aluno é automaticamente incrementado
        numero = indice_aluno
        indice_aluno += 1


        nomes_alunos.append(nome)
        numeros_alunos.append(numero)
        print("Aluno adicionado com sucesso!")

    elif escolha == "2":
        # Remover aluno
        if not nomes_alunos:
            print("Não existem alunos na turma.")
        else:
            numero = int(input("Insira o número do aluno que deseja remover: "))
            if numero in numeros_alunos:
                # Encontra o índice do número na lista dos números dos alunos
                indice = numeros_alunos.index(numero)
                # Remove o nome e o número do aluno correspondente nas listas
                del nomes_alunos[indice]
                del numeros_alunos[indice]
                print("Aluno removido com sucesso!")

                # Atualiza os índices após a remoção
                indice_aluno -= 1  # Atualiza o próximo índice

            else:
                print("Aluno não encontrado.")

    elif escolha == "3":
        # Consultar turma
        if not nomes_alunos:
            print("A turma está vazia.")
        else:
            print("Lista de Alunos:")
            for indice, (nome, numero) in enumerate(zip(nomes_alunos, numeros_alunos), start=1):
                # Imprime a lista de alunos enumerando-os
                print(f"Aluno {indice}: {nome} (Número: {numero})")

    elif escolha == "4":
        # Alterar nome do aluno
        if not nomes_alunos:
            print("A turma está vazia. Não há alunos para alterar.")
        else:
            numero = int(input("Insira o número do aluno cujo nome deseja alterar: "))
            if numero in numeros_alunos:
                # Encontra o índice do número na lista de números
                indice = numeros_alunos.index(numero)
                novo_nome = input("Insira o novo nome do aluno: ")
                # Atualiza o nome do aluno
                nomes_alunos[indice] = novo_nome
                print("Nome do aluno alterado com sucesso!")
            else:
                print("Aluno não encontrado.")

    elif escolha == "5":
        # Código para atualizar número do aluno
        if not nomes_alunos:
            print("A turma está vazia. Não há alunos para atualizar o número.")
        else:
            # Input para o utilizador introduzir o nº do Aluno a atualizar
            numero_antigo = int(input("Insira o número do aluno que deseja atualizar: "))
            if numero_antigo in numeros_alunos:
                # Encontra o índice do número na lista de números
                indice = numeros_alunos.index(numero_antigo)
                novo_numero = int(input("Insira o novo número do aluno: "))
                # Verifica se o novo número já existe
                if novo_numero in numeros_alunos:
                    print("Erro: Número de aluno já existente. Tente novamente.")
                else:
                    # Caso o nº não exista atualiza o nº de aluno
                    numeros_alunos[indice] = novo_numero
                    print("Número do aluno atualizado com sucesso!")
            else:
                print("Aluno não encontrado.")

    elif escolha == "6":
        # Código para limpar a tela (Google Colab)
        clear_output()

    elif escolha == "7":
        print("Programa Terminado! Obrigado!")
        break

    else:
        print("Opção inválida. Escolha uma opção de 1 a 7.")
```
