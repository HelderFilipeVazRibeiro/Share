```python
compras = []

while True:
  print("Selecione um das opções: ")
  print("1. Adicionar um item/produto na lista de compras")
  print("2. Consultar a lista de compras")
  print("3. Remover um item/produto da lista")
  print("4. Terminar Programa")

  opcao = input("Selecione a opção pretendida: ")

  if opcao == "1":
    item = input("Introduza o produto a adicionar à lista de compras: ")
    compras.append(item)
    print(f"O produto '{item}' foi adicionado à lista de compras. ")

  elif opcao == "2":
    if not compras:
      print("A lista de compras está vazia, por favor adicione um ou mais itens ")
    else:
      print("A lista contém estes itens/produtos: ")
      for i, item in enumerate(compras, start=1):
        print(f"{i} - {item}")

  elif opcao == "3":
     if not compras:
      print("A lista de compras está vazia, por favor adicione um ou mais itens ")
     else:
      try:
        indice = int(input("Introduza o id do item/produto que deseja remover da lista "))
        if 1 <= indice <= len(compras):
          item_removido = compras.pop(indice - 1)
          print(f"O item/produto '{item_removido}' foi removido da lista com sucesso! ")
        else:
          print("Por favor, certifique-se de que selecionou um id correto! ")

      except ValueError:
        print("Por favor, certifique-se de que introduziu um valor correto! ")

  elif opcao == "4":
    break

  else:
    print("Por favor, certifique-se de que a opção introduzida é valida! ")
```
